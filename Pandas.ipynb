**dropna()**

```python
import pandas as pd
import numpy as np   # ✅ added

data = {
    "name": ["ram","shyam","hari","rahul","aman","rohit"],
    "age": [12,14,15,16,17,18],
    "city": ["delhi","mumbai","up","assam","kerla","bihar"],
    "salary": [np.nan,40000,50000,60000,70000,80000]
}

df = pd.DataFrame(data)

print(df)
print(df.isnull())
df.dropna(axis=0,inplace=True)
print(df)
```

**replace null value with zero**

```python
import pandas as pd
import numpy as np   # ✅ added

data = {
    "name": ["ram","shyam","hari","rahul","aman","rohit"],
    "age": [12,14,15,16,17,18],
    "city": ["delhi","mumbai","up","assam","kerla","bihar"],
    "salary": [np.nan,40000,50000,60000,70000,80000]
}

df = pd.DataFrame(data)

print(df)
print(df.isnull())
df.fillna(0,inplace = True)
print(df)
```

**fill null with specific value**

```python
import pandas as pd
import numpy as np   # ✅ added

data = {
    "name": ["ram","shyam","hari","rahul","aman","rohit"],
    "age": [12,14,15,16,17,18],
    "city": ["delhi","mumbai","up","assam","kerla","bihar"],
    "salary": [np.nan,40000,50000,60000,70000,80000]
}

df = pd.DataFrame(data)

print(df)
print(df.isnull())
df["salary"].fillna(df['age'].mean(), inplace=True)
print(df)
```

**INTERPOSITION**(it puts the extimated value)

```python
import pandas as pd
import numpy as np
data = {
    "name": ["ram","shyam","hari","rahul","aman","rohit"],
    "age": [13,np.nan,15,np.nan,17,18],
    "city": ["delhi","mumbai","up","assam","kerla","bihar"],
    "salary": [np.nan,40000,50000,60000,70000,80000]
}
df = pd.DataFrame(data)
print(df)
print(df.isnull())
df["age"] = df["age"].interpolate(method="linear")
df["salary"].fillna(df["salary"].mean(), inplace=True)
print(df)
'''
three main type of interposition
-linear
-polynomial
-time
```

**sorting**

**Aggregation**(summary of the coloum )

```python
import pandas as pd
import numpy as np

data = {
    "name": ["ram","shyam","hari","rahul","aman","rohit"],
    "age": [13,31,15,26,17,18],
    "city": ["delhi","mumbai","up","assam","kerla","bihar"],
    "salary": [np.nan,40000,50000,60000,70000,80000]
}
df=pd.DataFrame(data)
print (df)
mmm = df["age"].mean()
sum = df["age"].sum()
min = df["age"].min()
max = df["age"].max()
print(mmm)
print(min)
print(max)
print(sum)
```

**grouping in pandas**

```python
import pandas as pd
import numpy as np

data = {
    "name": ["ram","shyam","hari","rahul","aman","rohit"],
    "age": [13,31,15,26,17,18],
    "city": ["delhi","mumbai","up","assam","kerla","bihar"],
    "salary": [np.nan,40000,50000,60000,70000,80000]
}
df=pd.DataFrame(data)
grouped = df.groupby(["age", "salary"]).sum()
print(grouped)
```

**merging and joining**

```python
import pandas as pd

df_customer = pd.DataFrame({
    "customer_id":[1,2,3,4,5],
    "name":["ram","shyam","hari","rahul","aman"],
})

df_order = pd.DataFrame({
    "customer_id":[1,2,3,4,5],
    "order_id":[101,102,103,104,105],
    "amount":[100,200,300,400,500],
})

df_merged = pd.merge(df_customer, df_order, on="customer_id", how="inner")

print(df_merged)

'''merger can be off
inner
outer
left
right'''
```
